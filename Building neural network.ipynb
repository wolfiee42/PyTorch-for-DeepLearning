```
import torch
import torch.nn as nn


# Create input tensor: 1 sample with 3 features
input_tensor = torch.tensor([[0.5, 0.8, 0.2]])
print("Input Shape:", input_tensor.shape) # torch.Size([1, 3])


# Create a linear layer: 3 inputs -> 2 outputs
linear_layer = nn.Linear(in_features=3, out_features=2)


# Access weights
print("Weight Shape:", linear_layer.weight.shape)
print("Weight:\n", linear_layer.weight)


# Access bias
print("\nBias Shape:", linear_layer.bias.shape)
print("Bias:", linear_layer.bias)


# Pass input through the layer
output = linear_layer(input_tensor)
print("output shape:", output.shape)  # torch.Size([1, 2])
print("Output:", output)
```

```
import torch
import torch.nn as nn


# Weather data: [temperature, humidity, wind_speed]
# Values are normalized (scaled between 0 and 1)
weather_data = torch.tensor([[0.5, 0.8, 0.2]])  # Hot, very humid, light wind

print("Input (weather features):", weather_data)
print("Shape:", weather_data.shape)


# Create network: 3 features -> 2 predictions
weathear_network = nn.Linear(in_features=3, out_features=2)


# Make prediction
prediction = weathear_network(weather_data)
print("\nRaw prediction (rain, clouds):", prediction)
```

```
import torch
import torch.nn as nn

torch.manual_seed(123)

# Step 1: Create input data (3 samples, 4 features each)
input_data = torch.tensor([
    [0.5, 0.3, 0.8, 0.2],
    [0.1, 0.9, 0.4, 0.6],
    [0.7, 0.2, 0.5, 0.9]
])

# Step 2: Create linear layer (4 inputs -> 2 outputs)
layer = nn.Linear(in_features=4, out_features=2)

# Step 3: Forward pass
output = layer(input_data)

# Step 4: Count parameters
total_params = sum(p.numel() for p in layer.parameters())

# Results
print("Input shape:", input_data.shape)   # [3, 4]
print("Output shape:", output.shape)      # [3, 2]
print("Total parameters:", total_params)  # 10
```
